```python
import numpy as np
import math

e_1 = 10**-3

alpha_1 = np.array([
    [0.32, -0.18,  0.02,  0.21],
    [0.16,  0.12, -0.14,  0.27],
    [0.37,  0.27, -0.02, -0.24],
    [0.12,  0.21, -0.18,  0.25] 
])

beta_1 = np.array([
     1.83,
    -0.65,
     2.23,
    -1.13 
])

Xo = np.array([
    0.01,
    0.02,
    0.03,
    0.04
])


```

```python
# function checking whether iteration process is convergent

def is_convergent(matrix):
    # creating list of sums of all values(converted to absolute) in each matrix column 
    abs_matrix = np.abs(matrix)
    column_sums = np.sum(abs_matrix, axis=0)
    sums_list = column_sums.tolist()
    
    print(sums_list)
    print(max(sums_list))
    # our condition of convergence
    max_sum = np.max(sums_list)
    if max_sum < 1:
        return True
    return False
  
is_convergent(alpha_1)

```

```python

# finding second norm of matrix
def matrix_second_norm(matrix):
    abs_matrix = np.abs(matrix)
    column_sums = np.sum(abs_matrix, axis=0)
    sums_list = column_sums.tolist()
    max_sum = np.max(sums_list)
    return max_sum
 
# print(matrix_second_norm(alpha_1))

# find "k" parameter (number of iterations needed to find roots with given precision)

def find_iterations_count(alpha_norm, beta_norm, init_approx_norm, precision):
    return int(math.log((precision * (1 - init_approx_norm)) / beta_norm, alpha_norm) - 1)

# def find_iterations_count_2(alpha_norm, beta_norm, init_approx_norm, precision):
#     return (math.log10((precision * (1 - init_approx_norm))))


# print(find_iterations_count(0.76, 0.54, 0.54, 0.0001))
iter_count_1 = (find_iterations_count(
    alpha_norm=matrix_second_norm(alpha_1),
    beta_norm=matrix_second_norm(beta_1),
    init_approx_norm=matrix_second_norm(Xo),
    precision=e_1
))
print("theoretical iteration count: ", iter_count_1)

```

```python

i = 0
X_next = np.array([
    0.01,
    0.02,
    0.03,
    0.04
])

while (i < iter_count_1):
    X_next = np.add(beta_1, np.dot(alpha_1, X_next))
    print(f"iteration {i + 1}: {X_next} ")
    i += 1
    
print(X_next)
```
