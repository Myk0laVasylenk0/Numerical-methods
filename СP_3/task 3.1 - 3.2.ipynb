### Завдання 3.1. 
#### Для матриці (табл. 3.1) знайдіть LU-розклад матриці.


```python
import numpy as np # useful library for maths operations
```

```python
# function that will perform LU decomposition of given matrix

def lu_decomposition(matrix):
    n = len(matrix)
    L = np.zeros((n, n))
    U = np.zeros((n, n))

    for i in range(n):
        L[i][i] = 1
        for j in range(i, n):
            sum_up = sum(L[i][k] * U[k][j] for k in range(i))
            U[i][j] = matrix[i][j] - sum_up

        for j in range(i + 1, n):
            sum_low = sum(L[j][k] * U[k][i] for k in range(i))
            L[j][i] = (matrix[j][i] - sum_low) / U[i][i]

    return L, U

# Example usage
A = np.array([[0.17, 0.75, -0.18, 0.21],
              [0.75, 0.13, 0.11, 1.00],
              [-0.33, 0.11, 3.01, -2.01],
              [0.11, 1.12, 1.11, -1.31]])

B = np.array([[0.11], 
              [2.00], 
              [0.11], 
              [0.13]])


L, U = lu_decomposition(A)
print("L:")
print(L)
print("U:")
print(U)
```

### Завдання 3.2. 
#### Використовуючи LU-розклад матриці системи (табл. 3.1), 
#### обчисліть її визначник, обернену матрицю та розв’язок системи.

```python
# Calculating the determinant of matrix A

L1 = np.copy(L)
U1 = np.copy(U)

detL = np.linalg.det(L1)
detU = np.linalg.det(U1)

detA = detL * detU
print("The determinant of matrix A:")
print(detA)
```

```python
# Calculating inverse matrix A

L2 = np.copy(L)
U2 = np.copy(U)

invL = np.linalg.inv(L2)
invU = np.linalg.inv(U2)

invA = np.dot(invU, invL)
print("Inverse matrix A:")
print(invA) 
```

```python
# Finding roots for matrix equasion 

invA1 = np.copy(invA)
X = np.dot(invA1, B)
print("Root matrix as a result of matrix equasion:")
print(X)
```
